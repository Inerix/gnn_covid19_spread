```python
import pandas as pd
import numpy as np
```

```python
# load and clean Canada confirmed cases data
dfc_can = pd.read_csv('novel-corona-virus-2019-dataset/time_series_covid_19_confirmed.csv')
dfc_can = dfc_can[dfc_can['Country/Region'] == 'Canada'].reset_index(drop=True)
dfc_can.drop(columns=['Lat', 'Long'], inplace=True)
dfc_can.head()
```

```python
# load and clean Canada deaths data
dfd_can = pd.read_csv('novel-corona-virus-2019-dataset/time_series_covid_19_deaths.csv')
dfd_can = dfd_can[dfd_can['Country/Region'] == 'Canada'].reset_index(drop=True)
dfd_can.drop(columns=['Lat', 'Long'], inplace=True)
dfd_can.head()
```

```python
# load and clean US confirmed cases data
us_column_drop = ['UID', 'iso2', 'iso3', 'code3', 'FIPS', 'Admin2', 'Lat', 'Long_',
                  'Combined_Key', 'Province_State', 'Country_Region']
dfc_us = pd.read_csv('novel-corona-virus-2019-dataset/time_series_covid_19_confirmed_US.csv')
dfc_us['Province/State'] = dfc_us['Province_State']
dfc_us['Country/Region'] = dfc_us['Country_Region']
dfc_us.drop(columns=us_column_drop, inplace=True)
dfc_us = dfc_us.groupby(['Province/State', 'Country/Region']).sum().reset_index()
dfc_us.head()
```

```python
# load and clean US deaths data
us_column_drop = ['UID', 'iso2', 'iso3', 'code3', 'FIPS', 'Admin2', 'Lat', 'Long_',
                  'Combined_Key', 'Population', 'Province_State', 'Country_Region']
dfd_us = pd.read_csv('novel-corona-virus-2019-dataset/time_series_covid_19_deaths_US.csv')
dfd_us['Province/State'] = dfd_us['Province_State']
dfd_us['Country/Region'] = dfd_us['Country_Region']
dfd_us.drop(columns=us_column_drop, inplace=True)
dfd_us = dfd_us.groupby(['Province/State', 'Country/Region']).sum().reset_index()
dfd_us.head()
```

```python
# concatenate data frames
dfc = pd.concat([dfc_us, dfc_can]).reset_index(drop=True)
dfd = pd.concat([dfd_us, dfd_can]).reset_index(drop=True)

# save data frames
dfc.to_pickle('cleaned_data/confirmed.pkl')
dfd.to_pickle('cleaned_data/deaths.pkl')
```
